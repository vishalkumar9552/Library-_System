```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

y_1 = sns.load_dataset("tips")
y_1
```

```python
sns.lineplot(x="total_bill",y="tip",data=y_1,hue="sex",size=10,style="sex",palette="Accent",markers=["o",">"],dashes=False
             ,legend=True)
plt.grid()
plt.title("PLOT B/W TOTAL_BILL,TIP AND SEX")
plt.show()
```

```python
sns.lineplot(x="smoker",y="day",data=y_1,hue="sex",size=10,style="sex",palette="Accent",markers=["o",">"],dashes=False
             ,legend=True)
plt.grid()
plt.title("PLOT B/W SMOKER,DATA AND SEX")
plt.show()
```

```python
sns.lineplot(x="day",y="time",data=y_1,hue="sex",size=10,style="sex",palette="Accent",markers=["o",">"],dashes=False
             ,legend=True)
plt.grid()
plt.title("PLOT B/W DAY,DATA AND SEX")
plt.show()
```
