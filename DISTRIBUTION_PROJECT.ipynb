```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n=10,p=0.5,size=1000),kde=True,)
plt.title('Binomial Distribution')
plt.show()

sns.displot(random.normal(loc=10,scale=4,size=1000),kde=True,)
plt.title('Normal Distribution')
plt.show()

sns.displot(random.exponential(scale=2,size=1000),kde=True,)
plt.title('Exponential Distribution')
plt.show()
sns.displot(random.uniform(low=0,high=10,size=1000),kde=True,)
plt.title('Uniform Distribution')
plt.show()
```
